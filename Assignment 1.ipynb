```python
# WMMDD THILAKARATHNE
# D/ENG/21/0062/EE
# ASSIGNMENT 1
# Github code  https://github.com/minsaradevin/Assignment-1-on-Intensity-transformation

import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

im = cv.imread('images/natasha_grayscale.jpg', cv.IMREAD_GRAYSCALE)
assert im is not None

#Creating the regions
t1 = np.linspace(0, 100, 101).astype('uint8')    
t2 = np.linspace(125, 249, 125).astype('uint8') 
t3 = np.linspace(225, 255,30).astype('uint8') 

#transform
transform = np.concatenate((t1,t2,t3), axis=0).astype('uint8')
#transform = np.concatenate((transform, t3), axis=0).astype('uint8')

#ploting the curve
fig, ax = plt.subplots()
fig.suptitle("Transformation curve")
ax.plot(transform)
ax.set_xlabel(r'Input, $f(\mathbf{x})$')
ax.set_ylabel(r'Output, $\mathrm{T}[f(\mathbf{x})]$')
ax.set_xlim(0,255)
ax.set_ylim(0,255)
ax.set_aspect('equal')
ax.grid()
plt.show()

image_transform = cv.LUT(im, transform)

fig, ax = plt.subplots(1,2, figsize=(10,20))
ax[0].imshow(im, cmap="gray")
ax[0].set_title("Original")
ax[1].imshow(image_transform, cmap="gray")
ax[1].set_title("Transformed")
plt.show()
```

```python

import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

f = cv.imread ('images/highlights_and_shadows.jpg', cv.IMREAD_COLOR)
assert f is not None

f_LAB = cv.cvtColor(f, cv.COLOR_BGR2LAB) 
L_f, q_f , r_f = cv.split(f_LAB)

gamma = .7
t = np.array([(i/255.)**gamma*255 for i in range (256)], np.uint8)
g = t[L_f]
plt.suptitle("Gamma correction Curve")
plt.plot(t)
plt.show()


fig, ax = plt.subplots(1,2, figsize=(10,3.5))
fig.suptitle("b.Original image and gamma corrected image")
ax[0].imshow(L_f, cmap="gray")
ax[0].set_title("Original")
ax[1].imshow(g,cmap="gray")
ax[1].set_title("Gamma Corrected")
plt.show()

plt.figure(figsize = [10, 5])
plt.subplot(1, 2, 1)
plt.gca().set_title('Original')
f_h = cv.calcHist([L_f],[0],None,[256],[0,256])
plt.plot(f_h)


plt.subplot(1, 2, 2)
plt.gca().set_title('Gamma Corrected')
g_h = cv.calcHist([g],[0],None,[256],[0,256])
plt.plot(g_h)
plt.show()


```

```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

f = cv.imread('Images/washed_out_aerial_image.tif', cv.IMREAD_GRAYSCALE)
assert f is not None

plt.figure(figsize=[20,10])
plt.subplot(1,2,1)
plt.gca().set_title('Original')
h = np.zeros(256)
h = [np.sum(f==i) for i in range (256)]
plt.bar(range(256),h)

plt.subplot(1,2,2)
plt.gca().set_title('Equalized')
e = cv.equalizeHist(f)
plt.hist(e.ravel(),256,[0,256])
plt.show()

fig, ax = plt.subplots(1, 2, figsize=(15,30))  # fixed subplot creation
ax[0].imshow(f, cmap="gray")
ax[0].set_title('Original')
ax[1].imshow(e, cmap='gray')
ax[1].set_title('Transformed')
plt.show()
```
